```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('311_monthlycomplaints_nta.csv')
df.head(5)
```

```python
df.dtypes
```

```python
df['date'] = pd.to_datetime(df.date)
```

```python
df.dtypes
```

```python
dfbk09 = df[df['ntacode']=='BK09']
```

```python
dfbk09.head(5)
```

```python
data = dfbk09.drop(['date','ntacode'], axis=1)
data.index = dfbk09.date
data.head(5)
```

```python
data1 = data[['air quality','animal abuse','appliance','asbestos','best/site safety','blocked driveway','borough office','broken parking meter','building/use','consumer complaint','dead or damaged tree','derelict vehicles','dirty conditions']]
```

```python
#creating the train and validation set
train = data1[:int(0.8*(len(data)))]
valid = data1[int(0.8*(len(data))):]
```

```python
from statsmodels.tsa.vector_ar.var_model import VAR
```

```python
model = VAR(endog=train)
model_fit = model.fit()
```

```python
prediction = model_fit.forecast(model_fit.endog, steps=len(valid))
```

```python
#converting predictions to dataframe
pred = pd.DataFrame(index=range(0,len(prediction)))
for j in range(0,13):
  for i in range(0, len(prediction)):
    pred.loc[i,j] = prediction[i,j]
```

```python
pred.head(5)
```

```python
pred.columns = ['air quality','animal abuse','appliance','asbestos','best/site safety','blocked driveway','borough office','broken parking meter','building/use','consumer complaint','dead or damaged tree','derelict vehicles','dirty conditions']
```

```python
from math import sqrt
from sklearn.metrics import mean_squared_error
```

```python
#check rmse
for i in pred.columns:
    print('rmse value for', i, 'is : ', sqrt(mean_squared_error(pred[i], valid[i])))
```

```python
pred.head(5)
```

```python
valid.head(5)
```

```python
plt.figure(figsize=(10, 5)) # set the size of the plot
for col in pred.columns:
    plt.plot(valid.index, pred[col], label='Predicted ' + col)
    plt.plot(valid.index, valid[col], label='Actual ' + col)
plt.title('Predicted vs. Actual Values')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```
